```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
```

```python
legal = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]
```

```python
#Undersampling
legal_sample = legal.sample(n=473)
legal_sample.shape
```

```python
df_undersample = pd.concat([legal_sample, fraud], ignore_index=True)
```

```python
df_undersample
```

```python
df_undersample.shape
```

```python
df_undersample.Class.value_counts()
```

```python
x = df_undersample.drop(columns=['Class'])
y = df_undersample['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print('Train set shape:', x_train.shape, y_train.shape)
print('\nTest set shape:', x_test.shape, y_test.shape)
```

```python
# Scaling the features
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.fit_transform(x_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scale, y_train)
```

```python
pred = model.predict(x_test_scale)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
